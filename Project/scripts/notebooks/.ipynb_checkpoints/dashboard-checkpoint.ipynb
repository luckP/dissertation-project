```python
# print time
def printTime(time):
    time = int(time*1000)
    hour = int(time/3600000)
    min = int((time - hour * 3600000) / 60000)
    sec = int(( time - hour * 3600000  - min*60000 )/1000)
    mil_sec = time - hour * 3600000  - min * 60000 - sec * 1000
    return format(hour, '02d') + ': ' + format(min, '02d') + ': ' + format(sec, '02d') + ': ' + format(mil_sec, '04d')

class PrintInfo:
    def __init__(self, data, info):
        self.current = -1
        self.data = data
        self.time = time_time.time()
        self.info = info

    def __iter__(self):
        return self

    def __next__(self):
        self.current += 1
        if self.current % int(1+len(self.data)/10000) == 0:
            s = str(self.current)+' | '
            s += "%.2f" % round((self.current /len(self.data))*100, 2) + '% | '
            s += printTime(time_time.time() - self.time) + ' | '
            s += printTime( (time_time.time() - self.time)*len(self.data)/(self.current+1) ) + ' | '
            s += 'Matching: ' + str(len(self.info)) + ' | '
            s += 'Ratio: ' + "%.2f" % round((len(self.info) /(self.current+1))*100, 2) + '% | '

            sys.stdout.write("\r" + s)
            sys.stdout.flush()

        if self.current < len(self.data):
            return self.data[self.current]
        raise StopIteration
        
df['year'] = [str(pd.Timestamp(time, unit='s').year) for time in df['time']]
df['year-month'] = [str(pd.Timestamp(time, unit='s').year) +'-'+ str(pd.Timestamp(time, unit='s').month) for time in df['time']]
df['year-month-day'] = [str(pd.Timestamp(time, unit='s').year) +'-'+ str(pd.Timestamp(time, unit='s').month) +'-'+ str(pd.Timestamp(time, unit='s').day) for time in df['time']]
df['weekday'] = [str(pd.Timestamp(time, unit='s').weekday()) for time in df['time']]
df['hour'] = [str(pd.Timestamp(time, unit='s').hour) for time in df['time']]
```

# Implement the algorithm  to simplify edges

## utilts function

```python
import math

def dotproduct(v1, v2):
    return sum((a*b) for a, b in zip(v1, v2))

def length(v):
    if math.sqrt(dotproduct(v, v)) == 0:
        return 0.00000001
    return math.sqrt(dotproduct(v, v))

def angle(v1, v2):
    return math.acos(dotproduct(v1, v2) / (length(v1) * length(v2)))

def vect(p):
    return [p[0][0] - p[1][0], p[0][1]-p[1][1]]

def intermediatePoints(l, num=4):
    p=[]
    for i in range(len(l[0])):
        p1 = l[0][i]
        p2 = l[1][i]
        dist = p2 - p1
        p.append([ p1+ i*dist/(num+1) for i in range(num+2)])
    return [[a, b] for a, b in zip(p[0], p[1])]
```

## Check line compatibility

```python
# four types of compatible lines

# angle
def angleCompatibility(l1, l2):
    return angle(vect(l1), vect(l2)) < np.pi/4

def scaleCompatibility(l1, l2):
    w_l1 = sqrt((l1[0][0] - l1[1][0])**2 + (l1[0][1]-l1[1][1])**2)
    w_l2 = sqrt((l2[0][0] - l2[1][0])**2 + (l2[0][1]-l2[1][1])**2)
    return (w_l1+w_l2)/abs( w_l1 - w_l2) < 0.13

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sklearn
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import scale

import sklearn.metrics as sm
from sklearn import datasets
from sklearn.metrics import confusion_matrix, classification_report
```

```python
iris = datasets.load_iris()

X = scale(iris.data)
y = pd.DataFrame(iris.target)
variable_names = iris.feature_names

X[:10, ]
```

## Building and dunning yout model

```python
clustering = KMeans(n_clusters=3, random_state = 5)
clustering.fit(X)
```

## Plot yout model outputs

```python
iris_df = pd.DataFrame(iris.data)
iris_df.columns = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width']
y.columns = ['Targets']
```

```python
color_theme = np.array(['darkgray', 'lightsalmon', 'powderblue', ''])

plt.subplot(1, 2, 1)
plt.scatter(x=iris_df['Petal_Length'], y=iris_df['Petal_Width'], c=color_theme[iris.target], s=50)
plt.title('Ground Truth Classification')

plt.subplot(1, 2, 2)
plt.scatter(x=iris_df.Petal_Length, y=iris_df.Petal_Width, c=color_theme[clustering.labels_], s=50)
plt.title('K-Means Classification')
```
