# Create Data

## Table 1

```python
#generate data
num_values = 10000
r = requests.get('http://names.drycodes.com/'+str(num_values))
varchar_vals = json.loads(r.text)
data_table1 = [ { 'val_varchar': varchar_vals[i], 'val_int': random.randint(0, i) } for i in range(num_values)]
```

```python
for d in data_table1:
    query = "INSERT INTO test_database.table1 (val_varchar, val_int) VALUES (%s, %s)"
    ## storing values in a variable
    values = (d['val_varchar'], d['val_int'])

    ## executing the query with values
    cursor.execute(query, values)

    ## to make final output we have to run the 'commit()' method of the database object
    db.commit()
```

## Table 2

```python
#generate data
num_values = 100000
r = requests.get('http://names.drycodes.com/'+str(num_values))
varchar_vals = json.loads(r.text)
data_table1 = [ { 'val_varchar': varchar_vals[i], 'table1_id': random.randint(1, 1000) } for i in range(num_values)]
```

```python
for d in data_table1:
    query = "INSERT INTO test_database.table2 (val_varchar, table1_id) VALUES (%s, %s)"
    ## storing values in a variable
    values = (d['val_varchar'], d['table1_id'])

    ## executing the query with values
    cursor.execute(query, values)

    ## to make final output we have to run the 'commit()' method of the database object
    db.commit()
```

# Search Data

```python
data = []
for d in range(1000):
    t = time()
    query = "SELECT * FROM test_database.table2 WHERE table1_id=" + str(d+1)
    
    ## getting records from the table
    cursor.execute(query)

    ## fetching all records from the 'cursor' object
    records = cursor.fetchall()
    for r in records:
        data.append([e for e in r])

data
    
```

```python
csv_file = "luc_csv.csv"
try:
    with open(csv_file, 'w') as f:
        for d in data:
            f.write("%s\n"%(str(d)[1:-1]))

except IOError:
    print("I/O error")


```
